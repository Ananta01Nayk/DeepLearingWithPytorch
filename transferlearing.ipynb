```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader,random_split
import matplotlib.pyplot as plt
```

```python
# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
#transform
transform = transforms.Compose([
    transforms.Resize((150,150)),
    transforms.ToTensor()]
)
```

```python
full_datset = datasets.ImageFolder(
    root='./PetImages',
    transform=transform
    )
```

```python
print(f'class : {full_datset.classes}')
print(f'lenght of full_dataset   { len(full_datset)}')
```

```python
# make the langth of tran and test size
train_size = int(0.8 * len(full_datset))
test_size = int(len(full_datset) - train_size)
```

```python
train_data, test_data = random_split(full_datset,[train_size,test_size])
```

```python
#load the data set
train_loader = DataLoader(train_data,batch_size=32,shuffle=True)
test_loader = DataLoader(test_data,batch_size=32,shuffle=True)
```
